```python
import polars.selectors as cs
```

```python
df_model = df.select(cs.numeric()).drop("id", "scrape_id", "host_id", "price_f", "standardized_pricef")
```

```python
df_model.columns
```

```python
for col in df_model.columns:
  if col != "Price":
    print(df_model.select(pl.corr(col, "Price", method="pearson")))
```

```python
df.filter(pl.col("Price") < df.select(pl.col("Price").quantile(0.95))).to_pandas()
```

```python
hist = px.histogram(df.filter(pl.col("Price") < df.select(pl.col("Price").quantile(0.95))).to_pandas(), x="Price",
                    nbins=50, title="Histogram of Price", opacity=0.75, color_discrete_sequence=["blue"], range_y=[0,1800])
hist.update_layout(bargap=0.1)
hist.show()
```

```python
bar = px.bar(df.group_by("room_type").agg(pl.col("room_type").len().alias("freq")).sort("freq", descending=True),
             x='freq', y="room_type", title="Tipo de Alojamiento", color='room_type',
             orientation='h')
bar.show()
```

```python
bar = px.bar(df.group_by("neighbourhood_cleansed").agg(pl.col("neighbourhood_cleansed").len().alias("freq")).sort("freq", descending=True).head(10),
             x='freq', y="neighbourhood_cleansed", title="Alcaldías con más alojamientos", color='neighbourhood_cleansed',
             orientation='h')
bar.show()
```

```python
sm = px.scatter_mapbox(df.filter(pl.col("Price") < df.select(pl.col("Price").quantile(0.95))).sample(n=5000, seed=42).to_pandas(),
                       lat='latitude', lon='longitude', color="Price", size="Price", hover_name='neighbourhood_cleansed',
                       color_continuous_scale=px.colors.sequential.Viridis_r, zoom=10)
sm.show()
```
